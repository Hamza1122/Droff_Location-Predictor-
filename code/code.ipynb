```python
df[['pick_up_time']]=pd.to_datetime(df['pick_up_time'],unit='s')
df.head(2)
```

```python
df['pick_date']=[d.date() for d in df['pick_up_time']]
df['pick_time']=[d.time() for d in df['pick_up_time']]
df.head(3)
```

```python
df[['Drop_Date','Drop_Time']]=df.pickup_datetime.str.split(expand=True)
df['pick_date']=[d.date() for d in df['pick_up_time']]
df['pick_time']=[d.time() for d in df['pick_up_time']]
```

```python
df=df.drop(['pick_up_time','pickup_datetime'],axis=1)
```

```python
df['Drop_Date']=pd.to_datetime(df['Drop_Date'])
df.head(3)
```

```python
df[['Drop_Year','Drop_Month','Drop_day']]=df['Drop_Date'].apply(lambda x:pd.Series(x.strftime("%Y,%m,%d").split(",")))
df[['Pick_Year','Pick_Month','Pick_day']]=df['pick_date'].apply(lambda x:pd.Series(x.strftime("%Y,%m,%d").split(",")))
df[['Pick Hour','Pick Minute']]=df['pick_time'].apply(lambda x:pd.Series(x.strftime("%H,%M").split(",")))

#df[['Drop_Hour','Drop_Minute']]=df['Drop_Time'].apply(lambda x:pd.Series(x.strftime("%H,%M").split(",")))

df[['Drop Hour','Drop Minute']]=df['Drop_Time'].str.split(':',expand=True)
#f['New','New1']=df['Drop_Time'].str.split(':').tolist()

df.head(3)
```

```python
df=df.drop(['Drop_Date','pick_date','pick_time','Drop_Time'],axis=1)
```

```python
df['drop_off']=data
df['pick_up']=pick_up
df['user_id']=user_id
df['pick_up_geohash']=pick_up_geohash
df['drop_off_geohash']=drop_off_geohash
print(df.head(3))

df['drop_off_geohash']
```
